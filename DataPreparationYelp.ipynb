```python
def stream_jsonl(filename):
    with open(filename, "r", encoding="utf-8") as file:
        for i, line in enumerate(file):
            yield msgspec.json.decode(line.strip())


# Convert streamed data to a Pandas DataFrame
biz = pd.DataFrame(stream_jsonl(
    "yelp_academic_dataset_business.json"))
print(biz.head())
```

# Data Preparation

```python
biz = biz[biz['is_open'] == 1]
biz.info()
```

```python
biz.groupby("state")["business_id"].count().sort_values(ascending=False)
```

```python
# Get businesses in most popular state
biz = biz[biz['state'] == "PA"]
biz = biz[['business_id', 'name', 'city', 'postal_code', 'categories', 'attributes', 'stars']]

# Get only restaurants
restaurants = biz[biz['categories'].str.contains(
    'Restaurant.*') == True].reset_index()
restaurants.head()

# Flatten all attributes into a list
all_attributes = [key for attr_dict in restaurants['attributes'].dropna()
                  for key in attr_dict.keys()]

# Count occurrences and get the most common attributes (adjust number as needed)
most_common_attrs = [attr for attr, count in Counter(
    all_attributes).most_common(10)]  # Top 10 attributes

# Create new feature columns based on most common attributes
for attr in most_common_attrs:
    restaurants[attr] = restaurants['attributes'].apply(lambda x: x.get(
        attr, None) if isinstance(x, dict) else None)

# Clean up Parking
restaurants['BusinessParking'] = restaurants['BusinessParking'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
restaurants['Parking'] = restaurants['BusinessParking'].apply(
    lambda x: ', '.join([k for k, v in x.items() if v]
                        ) if isinstance(x, dict) else None
)

# Clean up Alcohol
restaurants['Alcohol'] = restaurants['Alcohol'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Clean up RestaurantsPriceRange2
restaurants['RestaurantsPriceRange2'] = pd.to_numeric(
    restaurants['RestaurantsPriceRange2'], errors='coerce')

# Clean up Ambience
restaurants['Ambience'] = restaurants['Ambience'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

restaurants['Ambience_Type'] = restaurants['Ambience'].apply(lambda x: next(
    (k for k, v in x.items() if v), None) if isinstance(x, dict) else None)

restaurants = restaurants.drop(columns=['index','attributes', 'Ambience', 'BusinessParking'])

restaurants.head()
```

```python
print(restaurants["RestaurantsTakeOut"].unique())
print(restaurants["RestaurantsDelivery"].unique())
print(restaurants["BusinessAcceptsCreditCards"].unique())
print(restaurants["OutdoorSeating"].unique())
print(restaurants["RestaurantsReservations"].unique())
print(restaurants["HasTV"].unique())
print(restaurants["Alcohol"].unique())
print(restaurants["Parking"].unique())
print(restaurants["Ambience_Type"].unique())
```

```python
user_business_interaction = reviews.groupby(['user_id', 'business_id']).agg(
    # Number of times a user interacted with a business
    frequency=('stars', 'count'),
    # Average rating given by the user to the business
    avg_rating=('stars', 'mean')
).reset_index()

filtered_business_ids = set(restaurants['business_id'])
# Filter for interactions that are restaurants
user_business_interaction = user_business_interaction[
    user_business_interaction['business_id'].isin(filtered_business_ids)
]
user_business_interaction.reset_index(drop=True, inplace=True)

user_business_interaction.head()
```
